```python
date_count = df.groupby("date").count()
date_count.plot.line(y=['txn_description'],c='k', ylabel=('Number of Transactions'),title=('Number of Transactions per Day'))
plt.legend().remove()
```

```python
df['timestamp'] = pd.to_datetime(df['extraction'])
df['time']= df['timestamp'].dt.time.astype(str).str.slice(0,2)
# df['time'].astype(str).str.slice(0,2)
time_count = df.groupby("time").count()
time_count.plot.line(y=['txn_description'],c='k',ylabel='Number of Transactions',xlabel='Time',title='Transactions per Hour',legend=False)
```

```python
df['transactions per dow']= df['date'].dt.dayofweek
day_count = df.groupby('transactions per dow').count().rename(index={0: "Monday",1: "Tuesday",2: "Wednesday",3: "Thursday",4: "Friday",5: "Saturday",6: "Sunday"})
day_count.plot.line(y=['txn_description'],ylabel='Number of Transactions',xlabel='Day of the Week',title='Transactions per Day of the Week',legend=False,figsize = (8,5),c='k').set_ylim(ymin=0)

```
